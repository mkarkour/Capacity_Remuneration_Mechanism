```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load the data
df = pd.read_csv(r"..\data\co2-emission-intensity-15 +GDP.csv")

# Select variables for clustering
features = ['Clean GW', 'Fossil GW', 'Renewables GW', 'CO2 intensity gCO2/kWh']
X = df[features]

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Plotting settings
sns.set(style="whitegrid")
#palette = sns.color_palette("pastel")

# Scatter plot for clustering results
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Clean GW', y='CO2 intensity gCO2/kWh', hue='Cluster')
plt.title('Clustering of Countries by Clean Energy Production and CO2 Intensity')
plt.xlabel('Clean Energy Production (GW)')
plt.ylabel('CO2 Intensity (gCO2/kWh)')
plt.legend(title='Cluster')
plt.show()
```
