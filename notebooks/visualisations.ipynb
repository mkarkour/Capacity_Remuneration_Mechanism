# Visualisation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px


df = pd.read_csv(r"..\data\co2-emission-intensity-15 +GDP.csv")
```

```python
fig = px.pie(df, values='Clean GW', names='Country', title='Répartition des types de production d\'énergie')
fig.show()
```

```python
# Aggregate data for the pie chart
energy_types = df[['Country', 'Clean GW', 'Fossil GW']].melt(id_vars=['Country'], var_name='Type', value_name='GW')

# Pie chart for energy production types
fig = px.pie(energy_types, values='GW', names='Type', title='Distribution of Energy Production Types',
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.show()
```

```python
sns.set(style="whitegrid")

# Line chart for clean energy production over years
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Clean GW', hue='Country')
plt.title('Evolution of Clean Energy Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Clean Energy Production (GW)')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
fig = px.bar(df, x='Country', y=['Clean GW', 'Fossil GW', 'Renewables GW'], 
             title='Energy Production by Type for Each Country', 
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.show()
```

```python
# Histogram for CO2 intensity distribution
plt.figure(figsize=(12, 6))
sns.histplot(df, x='CO2 intensity gCO2/kWh', kde=True)
plt.title('Distribution of CO2 Intensity Among Countries')
plt.xlabel('CO2 Intensity (gCO2/kWh)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot for renewable energy production vs CO2 intensity
fig = px.scatter(df, x='Renewables GW', y='CO2 intensity gCO2/kWh', color='Country', 
                 title='Renewable Energy Production vs CO2 Intensity', 
                 color_discrete_sequence=px.colors.qualitative.Pastel)
fig.show()
```

```python
# Choropleth map for renewable energy production in Europe
fig = px.choropleth(df, locations='Country', locationmode='country names', color='Renewables GW',
                    hover_name='Country', title='Renewable Energy Production in Europe')
fig.show()
```

```python
# Bar chart for comparison of wind and solar energy production and CO2 emissions
fig = px.bar(df, x='Country', y=['Wind and Solar GW', 'Wind and Solar mtCO2'], 
             title='Wind and Solar Energy Production and CO2 Emissions by Country', 
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.show()
```
